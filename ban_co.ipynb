```python
import turtle
import math
#Vị trí bắt đầu
x = -200
y = -200
#Kích thước bàn cờ
rong = 450
dai = 500

#Tạo nhân vật
nv = turtle.Turtle()

nv.penup()
nv.goto(x,y)
nv.pendown()

for i in range(2):
    nv.forward(rong)
    nv.left(90)
    nv.forward(dai)
    nv.left(90)

nv.pensize(4)
nv.penup()
nv.goto(x+20,y+20)
nv.pendown()

for i in range(2):
    nv.forward(rong-40)
    nv.left(90)
    nv.forward(dai-40)
    nv.left(90)

nv.pensize(1)
nv.penup()
nv.goto(x+25,y+25)
nv.pendown()

for i in range(2):
    nv.forward(rong-50)
    nv.left(90)
    nv.forward(dai-50)
    nv.left(90)

nv.penup()
nv.goto(x+25,y+225)
nv.pendown()

for i in range(2):
    nv.forward(rong-50)
    nv.right(90)
    nv.forward(50)
    nv.right(90)    
    nv.forward(rong-50)
    nv.left(90)
    nv.forward(50)
    nv.left(90)

for i in range(4):
    nv.forward(50)
    nv.left(90)
    nv.forward(rong-250)
    nv.right(90)
    nv.forward(50)
    nv.right(90)
    nv.forward(rong-250)
    nv.left(90)

nv.penup()
nv.goto(x+25,y+475)
nv.pendown()

for i in range(2):
    nv.forward(rong-50)
    nv.right(90)
    nv.forward(50)
    nv.right(90)    
    nv.forward(rong-50)
    nv.left(90)
    nv.forward(50)
    nv.left(90)

for i in range(4):
    nv.forward(50)
    nv.left(90)
    nv.forward(rong-250)
    nv.right(90)
    nv.forward(50)
    nv.right(90)
    nv.forward(rong-250)
    nv.left(90)

nv.backward(rong-50)

nv.penup()
nv.goto(x+175,y+25)
nv.pendown()

nv.left(45)
nv.forward(math.sqrt(2*100*100))
nv.right(45)
nv.backward(100)
nv.right(45)
nv.forward(math.sqrt(2*100*100))
nv.left(45)

nv.penup()
nv.goto(x+175,y+375)
nv.pendown()

nv.left(45)
nv.forward(math.sqrt(2*100*100))
nv.right(45)
nv.backward(100)
nv.right(45)
nv.forward(math.sqrt(2*100*100))
nv.left(45)
turtle.done()
```
